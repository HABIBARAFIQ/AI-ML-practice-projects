```python
#  feature engineering
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
# Date ,month,year are in object so we have to bring it to int
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
# now drop date of journey colm
df.drop('Date_of_Journey',inplace=True,axis=1)
```

```python
df.head()
```

```python
# if we do only split by : then the minute section will also capture the Date part like '01:10 22 Mar' for this for the min part '10 22 Mar' will appear
# Thats why we will use lambda function so that befor splitting by (:) we split by (" "),by this we can cut off the Date part
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_minute']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df['Arrival_minute']=df['Arrival_minute'].astype(int)
df['Arrival_hour']=df['Arrival_hour'].astype(int)
```

```python
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
# same for departure
df['Dep_hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_minute']=df['Dep_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_minute']=df['Dep_minute'].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
# Lets talk about categorical value
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].isnull().value_counts()
# only 1 null value
```

```python
df['Total_Stops'].mode()
# maximum number is 1 stop
```

```python
# if non-stop flight-price is low ,high stops,high price
# so we need to make every stop a label according to the high ness like non-stop->0,1 stop->1 .......in this way
# nan has get 1 as max is 1 stop
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, np.nan:1, '4 stops':4})
```

```python
df['Total_Stops'].isnull().value_counts()
```

```python
df.head(2)
```

```python
df.drop("Route",inplace=True,axis=1)
```

```python
df.head(2)
```

```python
df['Duration_hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_minute']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.head()
```

```python
df['Duration_minute'].isnull().value_counts()
```

```python
df['Duration_minute'] = df['Duration_minute'].fillna('0')
```

```python
df['Duration_hour'].unique()
```

```python
df['Duration_minute'].unique()
```

```python
mask = df['Duration_hour'].str.contains('m')
mask.value_counts()
```

```python
df.loc[mask, 'Duration_minute'] = (
    df.loc[mask, 'Duration_minute'] +
    df.loc[mask, 'Duration_hour'].str.split('m').str[0]
)
```

```python
df.loc[mask,'Duration_hour']='0'
```

```python
df['Duration_hour']=df['Duration_hour'].astype(int)
df['Duration_minute']=df['Duration_minute'].astype(int)
```

```python
df.info()
```

```python
df.drop('Duration',axis=1,inplace=True)
```
